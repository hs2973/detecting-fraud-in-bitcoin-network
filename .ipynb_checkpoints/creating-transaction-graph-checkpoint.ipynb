```python
import pandas as pd
```

```python
txin_df = pd.read_csv('trimmed_dataset/txin.txt', delimiter='\t', names=['txID', 'i', 'addrID', 'value'])
txlinks_df = pd.read_csv('trimmed_dataset/txlinks.txt', delimiter='\t', names=['txID', 'i', 'prev_txID', 'prev_i'])
```

```python
txin_df.head()
```

```python
txlinks_df.head()
```

```python
df = pd.merge(txin_df, txlinks_df, how='left', on=['txID', 'i'])
```

```python
df
```

```python
df.value = df.value * 10**-8
```

```python
df
```

```python
df[['prev_txID', 'txID', 'value']].to_csv('trimmed_dataset/tx.edgelist', sep=' ', header=False, index=False)
```
